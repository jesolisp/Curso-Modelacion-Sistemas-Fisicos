```python
import numpy as np
import sympy as sym
import matplotlib.pyplot as plt

from matplotlib import cm
from mpl_toolkits.mplot3d import Axes3D

# pip install ipympl
%matplotlib ipympl 
```

```python
a = 1/4
x1 = np.linspace(-4, 4, 50)
x2 = np.linspace(-2, 2, 50)

X1, X2 = np.meshgrid(x1, x2)

V = lambda x1, x2, a: (1/2)*x1**2 + (1/4)*a*x2**4
u = X1
v = a*X2**3
```

```python
fig, ax = plt.subplots(subplot_kw={"projection": "3d"}, figsize=(10,8))
surf = ax.plot_surface(X1, X2, V(X1,X2, a), cmap = cm.viridis, linewidth=0, antialiased=False)
ax.contour(X1, X2, V(X1,X2, a), cmap = cm.viridis, offset = -1)
ax.set_xlabel('x1')
ax.set_ylabel('x2')
ax.set_zlabel('V(x1,x2)')
plt.title('Función definida positiva')
plt.show()
```

```python
fig, ax = plt.subplots()
q = ax.quiver(X1, X2, u, v)
#ax.quiverkey(q, X = np.max(X1), Y = np.max(X2), U=1,
#             label='Quiver key, length = 10', labelpos='E')

plt.show()
```

```python

ax = plt.figure(figsize=(10,8)).add_subplot(projection='3d')
ax.quiver(X1, X2, V(X1, X2, a), u, v, 0, length=0.1, normalize=True)
```

```python
a = 2
x1 = np.linspace(-3*np.pi, 3*np.pi, 50)
x2 = np.linspace(-np.pi, np.pi, 50)

X1, X2 = np.meshgrid(x1, x2)

V = lambda x1, x2, a: a*(1 - np.cos(x1)) + (1/2)*x2**2
u = a*np.sin(X1)
v = X2
```

```python
fig, ax = plt.subplots(subplot_kw={"projection": "3d"}, figsize=(10,8))
surf = ax.plot_surface(X1, X2, V(X1,X2, a), cmap=cm.viridis,linewidth=0, antialiased=False)
ax.contour(X1, X2, V(X1,X2, a), cmap = cm.viridis, offset = -1)
ax.set_xlabel('x1')
ax.set_ylabel('x2')
ax.set_zlabel('V(x1,x2)')
plt.title('Función definida positiva')
plt.show()
```

```python
fig, ax = plt.subplots()
q = ax.quiver(X1, X2, u, v)
ax.quiverkey(q, X=0.3, Y=1.1, U=10,
             label='Quiver key, length = 10', labelpos='E')

plt.show()
```

```python
ax = plt.figure(figsize=(10,8)).add_subplot(projection='3d')

ax.quiver(X1, X2, V(X1,X2,a), u, v, 0, length=0.1, normalize=True)
```

```python
x1 = np.linspace(-10, 10, 50)
x2 = x1

X1, X2 = np.meshgrid(x1, x2)

Vp = lambda x1, x2: -2*(x1**2 + x2**2)

fig, ax = plt.subplots(subplot_kw={"projection": "3d"}, figsize=(10,8))
surf = ax.plot_surface(X1, X2, Vp(X1,X2), cmap=cm.viridis,linewidth=0, antialiased=False)
ax.contour(X1, X2, Vp(X1,X2), cmap = cm.viridis, offset = -1)
ax.set_xlabel('x1')
ax.set_ylabel('x2')
ax.set_zlabel('V(x1,x2)')
plt.title('Función definida negativa')
plt.show()
```
