Considere un sistema autónomo representado por la siguiente ecuación

\begin{equation}
 \label{eqn:sys_aut}
 \dot{x} = f(x),
\end{equation}

donde $f:D\leftarrow R^{n}$ es un mapa local de Lipschitz de un dominio $D\subset R^{n}$ en $R^{n}$. 

Para poder estudiar la estabilidad del sistema \eqref{eqn:sys_aut}, es necesario obtener $\bar{x}\in D$ que denota el punto de equilibrio 

\begin{equation}
 f(\bar{x}) = 0.
\end{equation}

## Localización de los puntos de equilibrio

**Definición 1.** *El punto de equilibrio $x=0$ de $\dot{x} = f(x)$ es*

* *estable si, para cada $\varepsilon >0$, hay un $\delta = \delta(\varepsilon)>0$ tal que*

\begin{equation}
 \|x(0)\| < \delta \Rightarrow \|x(t)\| < \varepsilon, \quad \forall t\geq 0
\end{equation}

* *inestable si no es estable*
* *asintóticamente estable si este es estable y $\delta$ se escoge tal que*

\begin{equation}
 \|x(0)\| < \delta \Rightarrow \lim \limits_{t\rightarrow \infty} x(t) = 0.
\end{equation}

  \dot{x}_{2} &= - \frac{g}{l}\sin(x_{1}) - \frac{k}{m}x_{2}.
 \end{aligned}
\end{equation}

## Linealización de sistemas en torno a un punto de equilibrio

Un sistema lineal invariante en el tiempo (LTI) de la forma

\begin{equation}
 \label{eqn:LTI}
 \dot{x} = Ax,
\end{equation}

tiene un punto de equilibrio en el origen. Este punto de equilibrio está aislado si y sólo si $\text{det}(A)\neq0$. Si $\text{det}(A)=0$, la matriz $A$ tiene un espacio nulo no trivial.

Las propiedades de estabilidad del origen pueden ser caracterizadas por las ubicaciones de los eigenvalores de la matriz $A$.

Recordemos que la solución al sistema $\dot{x}=Ax$ dada una condición inicial $x(0)$ está dada como sigue

\begin{equation}
 x(t) = \exp(At)x(0).
\end{equation}

Además, para cualquier matriz $A$ existe una matriz $P$ no singular que transforma $A$ en su forma de Jordan, *i.e.*

\begin{equation}
 P^{-1}AP = J = \text{block} \text{diag}\left[J_{1}, J_{2},\dots, J_{r} \right],
\end{equation}

donde $J_{i}$ es el bloque de Jordan asociado con el eigenvalor $\lambda_{i}$ de A.

La forma generalizada del bloque de Jordan, considerando un orden $m$, tiene la forma siguiente

\begin{equation}
 J_{i} = \begin{bmatrix}
  \lambda_{i} & 1 & 0 & \cdots & \cdots & 0 \\
  0 & \lambda_{i} & 1 & 0 & \cdots & 0 \\
  \vdots & & \ddots & & & \vdots \\
  \vdots & & & \ddots & & 0 \\
  \vdots & & & & \ddots & 1 \\
  0 & \cdots & \cdots & \cdots & 0 & \lambda_{i}
 \end{bmatrix}_{m\times m},
\end{equation}

por consiguiente

\begin{equation}
 \exp(At) = P \exp(Jt)P^{-1} = \sum_{i=1}^{r} \sum_{k=1}^{m_{i}} t^{k-1} \exp(\lambda_{i}t) R_{ik},
\end{equation}

donde $m_{i}$ es el orden del bloque de Jordan asociado con el eigenvalor $\lambda_{i}$.

**Teorema 1**. *El punto de equilibrio $x=0$ de $\dot{x} = Ax$ es estable si y sólo si todos los eigenvalores de $A$ satisfacen $\text{Re}(\lambda_{i})\leq 0$ y cada eigenvalor con $\text{Re}(\lambda_{i})= 0$ tiene un bloque de Jordan asociado de uno. El punto de equilibrio $x=0$ es (globalmente) asintóticamente estable sí y sólo sí todos los eigencalores de $A$ satisfacen $\text{Re}(\lambda_{i})< 0$*.

La demostración de este teorema se puede consultar en Khalil (2002).

### Linealización 

Suponga que las funciones $f_{1}$ y $f_{2}$ del sistema autónomo de segundo orden

\begin{equation}
 \label{eqn:second_order_aut}
 \begin{aligned}
  \dot{x}_{1} &= f_{1}(x_{1},x_{2}), \\
  \dot{x}_{2} &= f_{2}(x_{1},x_{2}), \\
 \end{aligned}
\end{equation}

son continuamente diferenciables y que además $p=(p_{1},p_{2})$ es un punto de equilibrio, el sistema \eqref{eqn:second_order_aut} se puede expandir mediante serie de Taylor en torno a $(p_{1},p_{2})$ como sigue

\begin{equation}
 \begin{aligned}
  \dot{x}_{1} &= f_{1}(p_{1},p_{2}) + a_{11}(x_{1}-p_{1}) + a_{12}(x_{2}-p_{2}) + \text{H.O.T}, \\
  \dot{x}_{2} &= f_{2}(p_{1},p_{2}) + a_{21}(x_{1}-p_{1}) + a_{22}(x_{2}-p_{2}) + \text{H.O.T},
 \end{aligned}
\end{equation}

donde

\begin{equation}
 \begin{aligned}
  a_{11} = \left. \frac{\partial f_{1}(x_{1},x_{2})}{\partial x_{1}} \right|_{x_{1}=p_{1},x_{2}=p_{2}}, \quad a_{12} = \left. \frac{\partial f_{1}(x_{1},x_{2})}{\partial x_{2}} \right|_{x_{1}=p_{1},x_{2}=p_{2}}, \\
  a_{21} = \left. \frac{\partial f_{2}(x_{1},x_{2})}{\partial x_{1}} \right|_{x_{1}=p_{1},x_{2}=p_{2}}, \quad a_{22} = \left. \frac{\partial f_{2}(x_{1},x_{2})}{\partial x_{2}} \right|_{x_{1}=p_{1},x_{2}=p_{2}}, 
 \end{aligned}
\end{equation}

y H.O.T. denota los términos de orden superior de la expansión de Taylor, es decir $(x_{1}-p_{1})^{2},(x_{2}-p_{2})^{2},(x_{1}-p_{1})\times (x_{2}-p_{2})$, y así sucesivamente.

Dado que $(p_{1},p_{2})$ denota el punto de equilibrio, tenemos

\begin{equation}
 f_{1}(p_{1},p_{2}) = f_{2}(p_{1},p_{2}) = 0.
\end{equation}

Puesto que consideramos trayectorias cercanas a este punto, definimos 

\begin{equation}
 y_{1} = x_{1} - p_{1}, \quad y_{2} = x_{2} - p_{2}.
\end{equation}

Por consiguiente, las ecuaciones de estado quedan de la siguiente forma

\begin{equation}
 \begin{aligned}
  \dot{y}_{1} &= \dot{x}_{1} = a_{11}y_{1} + a_{12}y_{2} + \text{H.O.T}, \\
  \dot{y}_{2} &= \dot{x}_{2} = a_{21}y_{1} + a_{22}y_{2} + \text{H.O.T}.
 \end{aligned}
\end{equation}

Si consideramos una vecindad cercana al punto de equilibrio, entonces podemos decir que los términos H.O.T son despreciables y por consiguiente, realizamos una aproximación de la ecuación de estado no lineal por la ecuación de estado lineal como sigue

\begin{equation}
 \begin{aligned}
  \dot{y}_{1} &= \dot{x}_{1} = a_{11}y_{1} + a_{12}y_{2},\\
  \dot{y}_{2} &= \dot{x}_{2} = a_{21}y_{1} + a_{22}y_{2}. \\
 \end{aligned}
\end{equation}

Reescribiendo lo anterior en una forma vectorial, tenemos

\begin{equation}
 \dot{y} = Ay,
\end{equation}

donde

\begin{equation}
 A = \begin{bmatrix}
  a_{11} & a_{12} \\
  a_{21} & a_{22}  
 \end{bmatrix} =
 \left.
  \begin{bmatrix}
   \frac{\partial f_{1}}{\partial x_{1}} & \frac{\partial f_{1}}{\partial x_{2}} \\
   \frac{\partial f_{2}}{\partial x_{1}} & \frac{\partial f_{2}}{\partial x_{2}}
  \end{bmatrix} \right|_{x=p} =
  \left.
   \frac{\partial f}{\partial x}
  \right|_{x=p}
\end{equation}

**Observación.** La matrix $[\partial f/\partial x]$ es llamada la matriz Jacobiana de $f(x)$ mientras que $A$ es la matriz Jacobiana evaluada en el punto de equilibrio $p$.

## Análisis de la estabilidad de sistemas dinámicos linealizados

Considere un sistema LTI como se muestra en la Ec. \eqref{eqn:LTI} donde $A$ es una matriz real de dimensión $2\times 2$. La solución a este sistema con una condición inicial $x_{0}$ está dada por la siguiente ecuación

\begin{equation}
 x(t) = M \exp(J_{r}t)M^{-1}x_{0},
\end{equation}

donde $J_{r}$ es la forma real de Jordan de $A$ y $M$ es una matriz real no singular tal que $M^{-1}AM = J_{r}$. Dependiendo de los eigenvalores $\lambda_{i}$ de la matriz $A$, la forma de Jordan puede tomar los siguientes casos

\begin{equation}
 \begin{bmatrix}
  \lambda_{1} & 0 \\
  0 & \lambda_{2}
 \end{bmatrix}, \quad
 \begin{bmatrix}
  \lambda & k \\
  0 & \lambda
 \end{bmatrix}, \quad
 \begin{bmatrix}
  \alpha & -\beta \\
  \beta & \alpha
 \end{bmatrix},
\end{equation}

donde $k$ puede tomar el valor de $0$ o $1$.

### Caso 1. Ambos eigenvalores son reales: $\lambda_{1} \neq \lambda_{2} \neq 0$.

El sistema tiene dos eigenvectores reales $v_{1}$ y $v_{2}$ asociados con $\lambda_{1}$ y $\lambda_{2}$, respectivamente. Donde el sistema es transformado en dos ecuaciones diferenciales de primer orden después de un cambio de coordenadas $z=M^{-1}x$, *i.e.*

\begin{equation}
 \begin{aligned}
  \dot{z}_{1} &= \lambda_{1}z_{1},\\
  \dot{z}_{2} &= \lambda_{2}z_{2},
 \end{aligned}
\end{equation}

cuya solución está dada como sigue, para las condiciones iniciales $(z_{10},z_{20})$

\begin{equation}
 z_{1}(t) = z_{10}e^{\lambda_{1}t}, \quad z_{2}(t) = z_{20}e^{\lambda_{2}t}.
\end{equation}

Cuando $\lambda_{2}<\lambda_{1}<0$, el retrato fase tiene la forma de la Figura 1.15(a), donde $x=0$ es llamado *nodo estable*. Por otro lado, si $\lambda_{2}>\lambda_{1}>0$, $x=0$, el retrato fase se muestra en la Figura 1.15(b) y se le conoce como *nodo inestable*.

![Nodo estable e inestable](../figures/nodo_estable_inestable.png)

Si el sistema tiene eigenvalores con signos opuestos, esto es $\lambda_{2} < 0 \lambda_{1}$, decimos que $\lambda_{2}$ es un eigenvalor estable y $\lambda_{1}$ inestable. Por consiguiente, el retrato fase que corresponde a esta trayectoria se muestra en la Figura 1.16(a).

![Punto silla](../figures/punto_silla.png)

### Caso 2. Eigenvalores complejos: $\lambda_{1,2} = \alpha \pm j\beta$

El sistema $\dot{x}=Ax$ después de la transformación de coordenadas $z = M^{-1}x$ tiene la forma

\begin{equation}
 \begin{aligned}
  \dot{z}_{1} &= \alpha z_{1} - \beta z_{2}, \\
  \dot{z}_{2} &= \beta z_{1} + \alpha z_{2}.
 \end{aligned}
\end{equation}

Puesto que la solución a este sistema es oscilatoria, se puede expresar en coordenadas polares como sigue

\begin{equation}
 r = \sqrt{z_{1}^{2} + z_{2}^{2}}, \quad \theta = \tan^{-1}\left( \frac{z_{2}}{z_{1}} \right),
\end{equation}

teniendo dos ecuaciones diferenciales de primer orden desacopladas

\begin{equation}
 \begin{aligned}
  \dot{r} &= \alpha r, \\
  \dot{\theta} = \beta.
 \end{aligned}
\end{equation}

La solución de este sistema, para una condición inicial $(r_{0},\theta_{0})$ es

\begin{equation}
 r(t) = r_{0}r^{\alpha t}, \quad \theta(t) = \theta_{0} + \beta_{t}.
\end{equation}

Dependiendo del valor de $\alpha$, el retrato fase puede tomar alguna de las siguientes formas

**Observación.** El punto de equilibrio $x=0$ es un *foco estable* si $\alpha<0$, *foco inestable* si $\alpha>0$ y *centro* si $\alpha=0$ como se muestra a continuación

\end{equation}

## Análisis de estabilidad por el método directo de Lyapunov

Aleksandr Mikhailovich Lyapunov en 1892 presentó dos métodos para determinar la estabilidad de sistemas dinámicos representados en ecuaciones diferenciales ordinarias (ODEs).

El primer método utiliza la forma explícita de la solución de las ODEs para realizar el análisis de estabilidad mientras que el segundo método determina la estabilidad del sistema sin resolver las ecuaciones de estado.

Sea $V:D\rightarrow R$ una función continuamente diferenciable definida en un dominio $D\subset R^{n}$ que contiene el origen, entonces la derivada de $V$ sobre la trayectoria del sistema autónomo $\dot{x}=f(x)$, denotada por $\dot{V}(x)$ está dada como sigue

\begin{equation}
 \begin{aligned}
  \dot{V}(x) &= \sum_{i=1}^{n} \frac{\partial V}{\partial x_{i}} \dot{x}_{i} = \sum_{i=1}^{n} \frac{\partial V}{\partial x_{i}}f_{i}(x), \\
  &= \begin{bmatrix}
   \frac{\partial V}{\partial x_{1}}, & \frac{\partial V}{\partial x_{2}}, & \dots, & \frac{\partial V}{\partial x_{n}}
  \end{bmatrix}
  \begin{bmatrix}
   f_{1}(x) \\
   f_{2}(x) \\
   \vdots \\
   f_{n}(x)
  \end{bmatrix} = 
  \frac{\partial V}{\partial x}f(x).
 \end{aligned}
\end{equation}

Normalmente las funciones utilizadas son funciónes reales de variable vectorial. También es posible utilizar funciones integrales pero nos vamos a centrar en las primeras.

**Definición 1.** Una función $V$ es *definida positiva* si $V(x)>0 ~ \forall x\neq 0$ y $V(0)=0$.

**Ejemplos**
* $V(x) = x_{1}^{2} + 2x_{2}^{2}$
* $V(x) = x_{1}^{2} + \frac{2x_{2}^{2}}{1+x_{2}^{2}}$

Si reemplazamos cualquier valor de $x_{1}$ y $x_{2}$ entonces la función es positiva excepto en el origen.

**Definición 2.** Una función $V$ es *definida negativa* si $V(x)<0 ~ \forall x\neq 0$ y $V(0)=0$.

**Ejemplo**
* $V(x) = - x_{1}^{2} - \left(3x_{1}+2x_{2}\right)^{2}$

**Definición 3.** Una función $V$ es *semidefinida positiva* si $V(x)\geq0 ~ \forall x$ y $V(x)=0$.

**Ejemplo**
* $V(x) = \left(x_{1} + x_{2} \right)^{2}$.

**Definición 4.** Una función $V$ es *semidefinida negativa* si $-V(x)$ es semidefinida positiva, *i.e.* $-V(x)\geq0 ~ \forall x$ y $V(x)=0$

**Ejemplo**
* $V(x) = -\left(x_{1} + x_{2}\right)^{2}$

**Observación.** $\dot{V}(x)$ es diferente para cada sistema y además, si $\Phi(t;x)$ es la solución al sistema $\dot{x}=f(x)$ con valor inicial $x$ en $t=0$, entonces

\begin{equation}
 \dot{V}(x) = \frac{\mathrm{d}}{\mathrm{d}t}V\left.\left(\Phi(t;x)\right)\right|_{t=0}.
\end{equation}

### Principio del Teorema de Estabilidad de Lyapunov

Un sistema es estable si su $E$ -energía- total (una función definida positiva) es continuamente decreciente (la derivada en el tiempo de la $E$ total es definida negativa) hasta que alcanza un estado de equilibrio.

En otras palabras, un sistema es estable si su $E$ total se va perdiendo con el paso del tiempo.

**Teorema 2.** *Sea $x=0$ un punto de equilibrio para el sistema $\dot{x}=f(x)$ y $D\subset R^{n}$ un dominio que contiene $x=0$, entonces $V:D\rightarrow R$ es una función continuamente diferenciable tal que*

\begin{equation}
 V(0) = 0,
\end{equation}

\begin{equation}
 V(x) > 0 \text{ en } D-\{0\},
\end{equation}

\begin{equation}
 \dot{V}(x) \leq 0 \text{ en } D.
\end{equation}

*Entonces, $x=0$ es estable. Además, si*

\begin{equation}
 \dot{V}(x) <0 \text{ en } D-\{0\},
\end{equation}

*entonces $x=0$ es asintóticamente estable.*

La demostración de este teorema se puede consultar en Khalil (2002).

**Teorema 3.** *Sea $x=0$ un punto de equilibrio para el sistema $\dot{x} = f(x)$ y $V:R^{n}\rightarrow R$ una función continuamente diferenciable tal que*

\begin{equation}
 V(0) = 0, \quad V(x)>0, \quad \forall x \neq 0,
\end{equation}

\begin{equation}
 \|x\| \rightarrow \infty \Rightarrow V(x) \rightarrow \infty,
\end{equation}

\begin{equation}
 \dot{V}(x)<0, \forall x\neq 0,
\end{equation}

*entonces $x=0$ es globalmente asintóticamente estable*.
